1. What is the difference between a function and a method in Python?

•Difference between function and method in Python
A function is a block of code that performs a specific task and exists independently, while a method is a function that is associated with an object and defined within a class.

**Example:**
```python
# Function
def greet(name):
    return f"Hello, {name}"

    # Method (defined within a class)

                return f"Hello, {name}"
                ```

2. Explain the concept of function arguments and parameters in Python.

•Function arguments and parameters
Parameters are the variables listed in the function definition, while arguments are the actual values passed to the function when it's called.

**Example:**
```python
def add(a, b):  # a and b are parameters
    return a + b

result = add(3, 5)  # 3 and 5 are arguments
```

5. What are iterators in Python and how do they differ from iterables?

•Iterators vs Iterables
An iterable is any object that can return an iterator (like lists, tuples), while an iterator is an object that keeps state and produces the next value when `next()` is called.

**Example:**
```python
my_list = [1, 2, 3]  # iterable
my_iter = iter(my_list)  # iterator
next(my_iter)  # returns 1
```

6. Explain the concept of generators in Python and how they are defined.

•Generators
Generators are functions that yield values one at a time using the `yield` keyword, maintaining their state between calls.

**Example:**
```python
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1
```

7. What are the advantages of using generators over regular functions?

•Advantages of generators
- Memory efficient (produce items one at a time)
- Can represent infinite streams
- Maintain state between calls
- Lazy evaluation

9. Explain the purpose and usage of the `map()` function in Python.

• `map()` function
Applies a given function to each item of an iterable and returns a map object.

**Example:**
```python
numbers = [1, 2, 3]
squared = map(lambda x: x**2, numbers)
```

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

•Difference between `map()`, `reduce()`, `filter()`
- `map()`: applies function to all items.
- `filter()`: selects items where function returns True.
- `reduce()`: applies function cumulatively to reduce sequence to single value.

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.




````python
•class Fibonacci:
    def __init__(self, limit):
        self.limit = limit
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.limit:
            raise StopIteration
        result = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return result
````

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

````python
•def powers_of_two(max_exp):
    exp = 0
    while exp <= max_exp:
        yield 2 ** exp
        exp += 1
````
